```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = []
with open('resultado.csv','r') as f:
    #lines = f.readlines()
    while line := f.readline():
        currentline = line.rstrip().split(",")
        data.append(currentline[:9])
```

```python
# Converter strings para float
data = np.array(data, dtype=float)

# Separar colunas
x = data[:, 0]  # Primeira coluna como eixo X
ys = data[:, 1:]  # As demais colunas como eixos Y

legenda = {
   0: "C2H6",
   1: "C2h5+",
   2: "C2H4",
   3: "H",
   4: "H2",
   5: "NO",
   6: "HNO"
}


# Criar o gráfico
plt.figure(figsize=(10, 6))

for i in range(ys.shape[1]):  # Percorrer as colunas do Y
    plt.plot(x, ys[:, i], label=f'{legenda[i]}')

# Personalizar o gráfico
plt.xlabel("Eixo X")
plt.ylabel("Valores")
plt.title("Gráfico de Dados")
plt.legend()
plt.grid(True)

# Exibir
plt.show()
```
